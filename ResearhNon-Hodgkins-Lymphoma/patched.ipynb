```python
import os
import cv2
```

```python
def patch_images(input_folder, output_folder, n):
    # Loop melalui setiap kelas dalam folder input
    for class_folder in os.listdir(input_folder):
        class_path = os.path.join(input_folder, class_folder)
        
        print(f"class folder : {class_folder}")
        print(f"input folder : {input_folder}")
        print(f"class path : {class_path}")
        print(f"====================================================")

        # Pastikan itu adalah direktori dan bukan file
        if os.path.isdir(class_path):
            print(f"\nProcessing class: {class_folder}")

            # Buat folder output untuk kelas tersebut
            output_class_folder = os.path.join(output_folder, class_folder)
            os.makedirs(output_class_folder, exist_ok=True)
            
            print(f"output class folder : {output_class_folder}\n")
            
            count = 0
            # Loop melalui setiap gambar dalam kelas
            for image_name in os.listdir(class_path):
                image_path = os.path.join(class_path, image_name)
                # count = count + 1
                # print(f"number : {count}")
                # print(f"image name : {image_name}")
                # print(f"image path : {image_path}\n")
                
                if os.path.isfile(image_path):
                    # Periksa apakah file tersebut memiliki ekstensi gambar yang diinginkan
                    if image_name.lower().endswith(('.tif')):
                        count += 1
                        print(f"number : {count}")
                        print(f"image name : {image_name}")
                        print(f"image path : {image_path}\n")
                    else:
                        print(f"Skipping non-image file: {image_name}")
                else:
                    print(f"Skipping directory or non-file: {image_name}")
                
                
                
                if image_path.endswith("tif"):
                    # Baca gambar
                    image = cv2.imread(image_path)

                    # Ambil dimensi gambar
                    height, width, dim = image.shape
                    print(f"width : {width}")
                    print(f"height : {height}")
                    print(f"dim : {dim}\n")

#                     # Hitung ukuran potongan (sub-gambar) yang diinginkan
                    sub_image_height = height // n
                    sub_image_width = width // n
        
                    print(f"sub image width {width} / {n} : {sub_image_width}")
                    print(f"sub image height {height} / {n} : {sub_image_height}")

#                     # Proses pemisahan gambar menjadi potongan (sub-gambar)
                    for i in range(0, n):
                        for j in range(0, n):
                            # Hitung koordinat untuk memotong gambar
                            start_row = i * sub_image_height
                            end_row = start_row + sub_image_height
                            start_col = j * sub_image_width
                            end_col = start_col + sub_image_width
                            
                            # print(f"baris : {i}, kolom : {j}")
                            # print(f"start row : {start_row}")
                            # print(f"end row : {end_row}")
                            # print(f"start col : {start_col}")
                            # print(f"end col : {end_col}\n")
                            

#                             # Potong gambar
                            sub_image = image[start_row:end_row, start_col:end_col]
                            # height, width, dim = sub_image.shape
                            # print(f"sub width : {width}")
                            # print(f"sub height : {height}")
                            # print(f"sub dim : {dim}\n")

                            sub_image_filename = f"{image_name[:-4]}_sub_image_{i}_{j}.tif"
                            # print(f"sub image filename : {sub_image_filename}\n")
                            sub_image_path = os.path.join(output_class_folder, sub_image_filename)
                            # print(f"sub_image_path : {sub_image_path}")
                            cv2.imwrite(sub_image_path, sub_image)

                            print(f"Saved sub-image: {sub_image_path}")
                print(f"\n\n")
        print(f"jumlah {class_folder} : {count}")
        print(f"--------------------------------------DONEEE{class_folder}--------------------------------------------\n\n")
    print(f"++++++++++++++++++++++++++++++++++++++++++DONEEE++++++++++++++++++++++++++++++++++++++++++++\n\n")    
```

```python
# Sebutkan banyak patch 
patches = [2]

# Ganti dengan path folder dataset input
input_dataset_folder = "/workspace/bengkelkodingudinus_gmail.com/azizu/lymphoma/"

for patch in patches :
    # Ganti dengan path folder output untuk menyimpan sub-gambar
    output_sub_image_folder = f"patchingFlip/patchingRaw_{patch}"
    
    # str(patch)
    # Pemanggilan fungsi untuk memisah gambar menjadi sub-gambar
    patch_images(input_dataset_folder, output_sub_image_folder, patch)
```
