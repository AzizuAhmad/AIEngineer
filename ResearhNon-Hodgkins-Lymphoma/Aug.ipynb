```python
def flip_and_save_images(input_folder, output_folder):
    # Ensure the output folder exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Loop through each class folder in the input folder
    for class_folder in os.listdir(input_folder):
        class_path = os.path.join(input_folder, class_folder)

        # Create separate output folders for each rotation angle
        output_folder_CLL = os.path.join(output_folder, "CLL")
        output_folder_MCL = os.path.join(output_folder, "MCL")
        output_folder_FL = os.path.join(output_folder, "FL")
        

        for folder in [output_folder_CLL,output_folder_MCL,output_folder_FL]:
            if not os.path.exists(folder):
                os.makedirs(folder)
                
        count = 0
        i = 1
        # Loop through each image in the class folder
        for image_name in os.listdir(class_path):
            image_path = os.path.join(class_path, image_name)
            if image_path.endswith("tif"):
                # Read the image using OpenCV
                original_image = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)

                # Check if the image was read successfully
                if original_image is None:
                    print(f"Error reading image: {image_path}")
                    continue  # Skip to the next image

                flip = cv2.flip(original_image, -1)
                
                if class_folder == "CLL":
                    cv2.imwrite(os.path.join(output_folder_CLL, f'{class_folder}_image{i}_flipX.tif'), flip)
                    # cv2.imwrite(os.path.join(output_folder_CLL, f'{class_folder}_image{i}.tif'), original_image)
                elif class_folder == "FL":
                    cv2.imwrite(os.path.join(output_folder_FL, f'{class_folder}_image{i}_flipX.tif'), flip)
                    # cv2.imwrite(os.path.join(output_folder_FL, f'{class_folder}_image{i}.tif'), original_image)
                elif class_folder == "MCL":
                    cv2.imwrite(os.path.join(output_folder_MCL, f'{class_folder}_image{i}_flipX.tif'), flip)
                    # cv2.imwrite(os.path.join(output_folder_MCL, f'{class_folder}_image{i}.tif'), original_image)
                
                i = i + 1
                                    
#                 count = count + 1
#                 if count == 1:
#                     break

if __name__ == "__main__":
    # Specify the input and output folders
    input_folder = "/workspace/bengkelkodingudinus_gmail.com/azizu/lymphoma/"
    output_folder = "lymphomaAug/JustFlip/x/"

    # Call the function to rotate and save images
    flip_and_save_images(input_folder, output_folder)
```

```python
dataset_dir = "patchingFlip/patchingRaw_2/"
```

```python
# List label
class_names = []

# Fitur dan label
features = []
labels = []

# Dimensi gambar
# Mengakses folder dataset
for path, subdirs, files in os.walk(dataset_dir):

    # Memilih jumlah dataset
    num_files = len(files)

    # Ganti 1 sesuai jumlah yang diinginkan (ex. 0.3)
    num_to_select = int(num_files * 1)
    selected_files = random.sample(files, num_to_select)

    # Mengakses file, resize, normalisasi, dan labeling
    for name in tqdm(selected_files):
        img_path = os.path.join(path, name)
        if img_path.endswith("tif"):
            image_read = cv2.imread(img_path)

            path_parts = path.split('/')
            label = path_parts[-1]

            if label not in class_names:
                class_names.append(label)

            features.append(image_read)
            index = class_names.index(label)
            labels.append(index)

# Meruabah fitur dan label menjadi array numpy
features = np.asarray(features)
labels = np.asarray(labels)
```
